```python
from skimage import io, img_as_ubyte

img = img_as_ubyte(io.imread("images/R1.jpg", as_gray = True))

print(img[10:15, 20:25])

mean_grey = img.mean()
max_value = img.max()
min_value = img.min()

print("Min, Max, and Mean are: ", min_value, max_value, mean_grey)
```

# Fliping and Plotting Image

```python
from scipy import ndimage
import numpy as np
from matplotlib import pyplot as plt

flippedLR = np.fliplr(img)
flippedUD = np.flipud(img)

plt.subplot(2,1,1)
plt.imshow(img, cmap="Greys")

plt.subplot(2,2,3)
plt.imshow(flippedLR, cmap="Blues")

plt.subplot(2,2,4)
plt.imshow(flippedUD, cmap="hsv")
```

# Fliters

```python
# Guassian fliter 

gaussian_filtered = ndimage.gaussian_filter(img,sigma=13)
plt.subplot(2,1,1)
plt.imshow(gaussian_filtered)

# Median filter
median_filtered = ndimage.median_filter(img,6)
plt.subplot(2,2,3)
plt.imshow(median_filtered)
```
