# Matplotlib 

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread("R1.jpg")
# print(type(img))
# print(img.shape)
plt.imshow(img)
plt.colorbar()
```

# scikit-image

```python
from skimage import io, img_as_float, img_as_ubyte
import matplotlib.pyplot as plt

image = io.imread("R1.jpg")
image_float = img_as_float(image)
print(image_float)
```

# Opencv

```python
import cv2
from matplotlib import pyplot as plt

grey_img = cv2.imread("R1.jpg", 0)
color_img = cv2.imread("R1.jpg", 1)

cv2.imshow("Grey Image", grey_img)
cv2.imshow("Color Image", color_img)

# Wait for a key event and close the OpenCV windows
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display the color image using Matplotlib
plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
plt.title("Color Image")
plt.show()
```
